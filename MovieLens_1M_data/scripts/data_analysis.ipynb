# 打分分布

```python
score_dic = {}
with open('./ratings.dat', 'r') as f:
    for line in f:
        info = line.strip()
        user, item, score, timestamp = info.split('::')
        if score not in score_dic:
            score_dic[score] = 0
        score_dic[score] += 1
print(score_dic)
```

```python
user_score_dic = {}

with open('./ratings.dat', 'r') as f:
    for line in f:
        info = line.strip()
        user, item, score, timestamp = info.split('::')
        if user not in user_score_dic:
            user_score_dic[user] = [0, 0]
        if float(score) >= 4:
            user_score_dic[user][1] += 1
        else:
            user_score_dic[user][0] += 1

for u in user_score_dic:
    if user_score_dic[u][1] == 1:
        print(u)
```

```python
import requests

# 配置
API_KEY = "40d4cf4cddf3821782fc4a1dbae820e5"
BASE_URL = "https://api.themoviedb.org/3"

def search_movie_by_title(title):
    url = f"{BASE_URL}/search/movie"
    params = {
        'api_key': API_KEY,
        'query': title,
        'language': 'en-US',      # 可改为 'zh-CN' 获取中文数据
        'page': 1,
        'include_adult': False
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
        data = response.json()
        if data['results']:
            return data['results'][0]  # 返回第一个匹配结果
        else:
            print("未找到电影")
            return None
    else:
        print(f"请求失败: {response.status_code}")
        return None

# 使用示例
movie = search_movie_by_title("Toy Story")
if movie:
    print(f"标题: {movie['title']}")
    print(f"上映年份: {movie['release_date']}")
    print(f"评分: {movie['vote_average']}/10")
    print(f"简介: {movie['overview']}")
    poster_path = movie['poster_path']
    if poster_path:
        poster_url = f"https://image.tmdb.org/t/p/w500{poster_path}"
        print(f"海报: {poster_url}")
```
